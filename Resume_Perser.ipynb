```python
import spacy
import re
import pandas as pd
import textract
from PIL import Image
import pytesseract
from difflib import SequenceMatcher

# Load the pre-trained SpaCy model for English
nlp = spacy.load("en_core_web_sm")

# Function to read skills from the text file
def read_skills(file_path):
    with open(file_path, 'r') as file:
        skills = file.read().split()
    return skills

# Function to extract candidate information from CV
def extract_info(cv_text, skills):
    # Define regular expression patterns for email and phone number extraction
    email_pattern = r"\b[A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Z|a-z]{2,}\b"
    phone_pattern = r"\b(?:\+\d{1,2}\s)?\d{3}[-.]?\d{3}[-.]?\d{4}\b"

    # Find email and phone number using regular expressions
    email_match = re.search(email_pattern, cv_text)
    phone_match = re.search(phone_pattern, cv_text)

    # Process the text using the SpaCy model
    doc = nlp(cv_text)

    # Initialize variables to store extracted information
    candidate_name = ""
    extracted_skills = []
    organizations = []
    education = []

    # Iterate over the entities recognized by SpaCy
    for ent in doc.ents:
        if ent.label_ == "PERSON" and not candidate_name:
            candidate_name = ent.text
        elif ent.label_ == "ORG":
            organizations.append(ent.text)
        elif ent.label_ == "EDUCATION":
            education.append(ent.text)

    # Extracted email and phone number
    email = email_match.group() if email_match else ""
    phone_number = phone_match.group() if phone_match else ""

    # Check if any skills from the list are present in the text with near match
    for skill in skills:
        if SequenceMatcher(None, skill.lower(), cv_text.lower()).ratio() >= 0.6:  # Adjust threshold as needed
            extracted_skills.append(skill)

    return candidate_name, email, phone_number, ", ".join(extracted_skills), ", ".join(organizations), ", ".join(education)

def process_cv(file_path, skills_file_path):
    # Read skills from the text file
    skills = read_skills(skills_file_path)

    # Read CV text based on file type
    if file_path.endswith(".pdf"):
        cv_text = textract.process(file_path).decode("utf-8")
    elif file_path.endswith((".docx", ".doc")):
        cv_text = textract.process(file_path).decode("utf-8")
    elif file_path.endswith((".jpg", ".jpeg", ".png")):
        cv_text = pytesseract.image_to_string(Image.open(file_path))
    else:
        raise ValueError("Unsupported file format")

    # Extract information from CV
    candidate_name, email, phone_number, skills, organizations, education = extract_info(cv_text, skills)

    # Create DataFrame
    df = pd.DataFrame({
        "Candidate Name": [candidate_name],
        "Email": [email],
        "Phone Number": [phone_number],
        "Skills": [skills],
        "Organizations Worked": [organizations],
        "Education": [education]
    })

    # Save DataFrame to Excel file
    df.to_excel("candidate_info.xlsx", index=False)
    print("Candidate information extracted and saved to candidate_info.xlsx")

# Provide the path to the CV file
cv_file_path = "/content/Sudipta_Mondal_Resume2024.docx"

# Provide the path to the skills file
skills_file_path = "skills.txt"

# Process the CV
process_cv(cv_file_path, skills_file_path)
```
